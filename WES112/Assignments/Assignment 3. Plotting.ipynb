### Exercise 1

Draw the following function over the interval [$x_1=\frac{\pi}{3} ; x_1=2\pi+\frac{\pi}{3}$] and make…
- the title of the drawing is “_A sawtooth function_”,
- the x- and y-axes are labelled as good as it can (Try to get “$x_1$” and “$y_1$”.),
- the length of a x-unit and a y-unit are the same,
- a grid appears.

$\displaystyle y_1 =  \pi - 2 \left(\sum_{i=1}^{20}\frac{\sin(ix_1)}{i} \right)$

Try it with different intervals for $x_1$ to become a smooth curve without computing too much points.

```octave
%plot --format svg
clear all
x1 = [-pi/3:0.05:7*pi/3] ;
i = [1:20] ;
A = i'*x1 ;
B = sin(A) ;
[xx,ii] = meshgrid(x1,i) ;
clear xx ;
C = B./ii ;
y1 = pi-2*sum(C) ;
plot(x1,y1) ;
title ("A sawtooth function")
xlabel("x_1")
ylabel("y_1")
axis equal
grid on
```
